```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 显示中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline
```

```python
# 导入数据
x=np.arange(0,10)
plt.title('这是一个示例标题')
plt.plot(x,x*x)
plt.show()
```

```python
# 添加文字
'''
text(x,y,text)
第一，二参数来设置坐标
第三个参数是设置显示文本内容
'''


x=np.arange(-10,11,1)
y=x*x
plt.plot(x,y)
plt.title('这是一个标题')
# 添加文字
plt.text(-2.5,30,'function y=x*x')
plt.show()
```

```python
# 添加注释
'''
annotate(xy,xytext,arrowprps)
xy参数:备注的坐标点
xytext参数：备注文字的坐标
arrowprops参数：在xy和xytext之间绘制一个箭头
'''

x=np.arange(-10,11,1)
y=x*x
plt.title('这是一个示例标题')
plt.plot(x,y)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'}))
plt.show()

```

```python
# 设置坐标轴名称
'''
xlabel()
ylabel()
'''
# 生成数组
x=np.arange(-10,11,1)
y=x*x
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
plt.plot(x,y)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
plt.show()
```

```python
# 添加图例
'''
legend()
'''
# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
plt.show()

```

```python
# 调整颜色
'''
plot中的color参数,样式有以下几种
color='g'
color='0.5'
color='#FF00FF'
color=(0.1,0.2,0.3)
'''
# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
plt.plot(x,y,color='g')
plt.plot(x,y1,color='0.5')
plt.plot(x,y2,color='#FF00FF')
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
plt.show()
```

```python
# 切换线条样式

'''
使用参数mark参数
‘.’：点(point marker)
‘,’：像素点(pixel marker)
‘o’：圆形(circle marker)
‘v’：朝下三角形(triangle_down marker)
‘^’：朝上三角形(triangle_up marker)
‘<‘：朝左三角形(triangle_left marker)
‘>’：朝右三角形(triangle_right marker)
‘1’：(tri_down marker)
‘2’：(tri_up marker)
‘3’：(tri_left marker)
‘4’：(tri_right marker)
‘s’：正方形(square marker)
‘p’：五边星(pentagon marker)
‘*’：星型(star marker)
‘h’：1号六角形(hexagon1 marker)
‘H’：2号六角形(hexagon2 marker)
‘+’：+号标记(plus marker)
‘x’：x号标记(x marker)
‘D’：菱形(diamond marker)
‘d’：小型菱形(thin_diamond marker)
‘|’：垂直线形(vline marker)
‘_’：水平线形(hline marker)

'''
# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
plt.plot(x,y,color='g',marker='o')
plt.plot(x,y1,color='0.5',marker='>')
plt.plot(x,y2,color='#FF00FF',marker='s')
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
plt.show()
```

```python
# 显示数学公式
'''
使用的是plt.text(x位置，y位置，公式，大小)
'''

# 设置坐标轴刻度范围
plt.xlim([1,8])
plt.ylim([1,5])

plt.text(2,4,r'$ \alpha \beta \pi \lambda \omega $',size=25)
plt.text(4,4,r'$ \sin(0)=\cos(\frac{\pi}{2}) $',size=25)
plt.text(2,2,r'$ \lim_{x \rightarrow y} \frac{1}{x^3} $',size=25)
plt.text(4,2,r'$ \sqrt[4]{x}=\sqrt{y} $',size=25)
plt.show()


```

```python
# 显示网格

'''
grid()
也可以设置颜色，线条的宽度，线条样式
grid(color='',linewith='',linestyle='')
'''

# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
plt.plot(x,y,color='g',marker='o')
plt.plot(x,y1,color='0.5',marker='>')
plt.plot(x,y2,color='#FF00FF',marker='s')
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
plt.grid()
plt.show()
```

```python
# 调整坐标轴刻度
'''
locator_params接口来调整
同时调整x轴与y轴
plt.locator_params(nbins=20)
只调整x轴：plt.locator_params('x',nbins=20)
只调整y轴：plt.locator_params('y',nbins=20)
'''

# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
# 绘制图形
plt.plot(x,y,color='g',marker='o')
plt.plot(x,y1,color='0.5',marker='>')
plt.plot(x,y2,color='#FF00FF',marker='s')
# 调整坐标轴刻度
plt.locator_params(nbins=20)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
# 设置网格线
plt.grid()
# 数据展示
plt.show()
```

```python
# 调整坐标轴范围
'''
axis:[0,5,0,10],x从0-5，y从0-10
xlim:对应参数有xmin,xmax，分别能调整最大值与最小值
ylim:同xlim用法
'''

# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
# 绘制图形
plt.plot(x,y,color='g',marker='o')
plt.plot(x,y1,color='0.5',marker='>')
plt.plot(x,y2,color='#FF00FF',marker='s')
# 设置坐标轴
plt.xlim(xmin=-20,xmax=20)
# 调整坐标轴刻度
plt.locator_params(nbins=20)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
# 设置网格线
plt.grid()
# 数据展示
plt.show()

```

```python
# 调整日期自适应
'''
plt.gcf().autofmt_xdate()
自动调整日期角度
'''
import pandas as pd

x=pd.date_range('2020/01/01',periods=30)
y=np.arange(0,30,1)
plt.plot(x,y)
# 日期自适应角度
plt.gcf().autofmt_xdate()
plt.show()
```

```python
# 添加双坐标轴-twinx

x=np.arange(1,20)
y1=x*x
y2=np.log(x)
plt.plot(x,y1)
# 添加一个坐标轴，从0到1
plt.twinx()
plt.plot(x,y2,'r')
plt.show()
```

```python
# 设置填充区域_fill()

x=np.linspace(0,5*np.pi,1000)
y1=np.sin(x)
y2=np.sin(2*x)
# 绘制图形
plt.plot(x,y1)
plt.plot(x,y2)
# 填充
plt.fill(x,y1,'g')
plt.fill(x,y2,'r')
plt.show()
```

```python
# 另一种填充的方式:fill_between(x,y1,y2,where=y1>y2,interpolate=True)
# 导入数据
x=np.linspace(0,5*np.pi,1000)
y1=np.sin(x)
y2=np.sin(2*x)
# 绘制图形
plt.plot(x,y1)
plt.plot(x,y2)
# 填充
plt.fill_between(x,y1,y2,where=y1>y2,interpolate=True)
plt.show()
```

```python
plt.style.use('ggplot')
# 生成数组
x=np.arange(-10,11,1)
y=x*x
y1=x*2
y2=x*3
# 设置坐标轴
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置标题
plt.title('这是一个示例标题')
# 绘制图形
plt.plot(x,y,color='g',marker='o')
plt.plot(x,y1,color='0.5',marker='>')
plt.plot(x,y2,color='#FF00FF',marker='s')
# 设置坐标轴
plt.xlim(xmin=-20,xmax=20)
# 调整坐标轴刻度
plt.locator_params(nbins=20)
# 添加注释
plt.annotate('这是一个示例注释',xy=(0,1),xytext=(-2,22),arrowprops={'headwidth':10,'facecolor':'r'})
# 添加图例
plt.legend(['示例1','示例2','示例3'])
# 设置网格线
plt.grid()
# 数据展示
plt.show()
```

```python
# 另一个绘制方式

plt.style.use('seaborn')

# 新建4个子图
fig,axes=plt.subplots(2,2)
ax1,ax2,ax3,ax4=axes.ravel()
# 绘制第一个子图图形
x,y=np.random.normal(size=(2,100))
ax1.plot(x,y,'o')
# 第二个子图
x=np.arange(0,10)
y=np.arange(0,10)
colors=plt.rcParams['axes.prop_cycle']
length=np.linspace(0,10,len(colors))
for s in length:
    ax2.plot(x,y+s,'-')
    
# 第三个图
x=np.arange(5)
y1,y2,y3=np.random.randint(1,25,size=(3,5))
width=0.25
ax3.bar(x,y1,width)
ax3.bar(x+width,y2,width)
ax3.bar(x+2*width,y3,width)


# 第四个图
for i,color in enumerate(colors):
    xy=np.random.normal(size=2)
ax4.add_patch(plt.Circle(xy,radius=0.3,color=color['color']))
ax4.axis('equal')
plt.show() 
```
