```python
# 导入水球图
from pyecharts.charts import Liquid,Grid

from pyecharts import options as opts
```

```python
# 基础图形展示

e=(
    Liquid()
    .add('水球图',[0.6,0.7])
    .set_global_opts(title_opts=opts.TitleOpts(title='水球图'))
)

e.render_notebook()
```

```python
# 多个图形展示

from pyecharts.commons.utils import JsCode
```

```python
c1=(
    Liquid()
    .add('水球图',[0.6,0.7],center=['60%','50%'])
    .set_global_opts(title_opts=opts.TitleOpts(title='多个水球图显示'))
)


c2=(
    Liquid()
    .add('水球图2',[0,3254],center=['25%','50%'],
    label_opts=opts.LabelOpts(font_size=50,formatter=JsCode(
            """function (param) {
                    return (Math.floor(param.value * 10000) / 100) + '%';
                }"""
        ),
    position='inside',),
)
)


# c1.render_notebook()
    
grid=Grid().add(c1,grid_opts=opts.GridOpts()).add(c2,grid_opts=opts.GridOpts())

grid.render_notebook()

```

```python
# 换个形状显示
# 更改样式
from pyecharts.globals import SymbolType
```

```python
c3=(
    Liquid()
    .add(
        "更换形状的水球图",
        [0.6,0.7],
        is_outline_show=False,
        # 样式转换
        shape=SymbolType.ARROW
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='更换图形的水球图'))
)

c3.render_notebook()
```
