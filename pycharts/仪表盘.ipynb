```python
from pyecharts.charts import  Gauge
from pyecharts import options as opts
```

```python
c=(
    Gauge()
    .add('',[('完成率',66.6)])
    .set_global_opts(title_opts=opts.TitleOpts(title='仪表盘示例'))

)

c.render_notebook()
```

```python
# 设置不同的颜色


c=(
    Gauge()
    .add(
        '业务指标',
        [('完成率',55.5)],
        axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(
        color=[(0.3, "#67e0e3"), (0.7, "#37a2da"), (1, "#fd666d")], width=30
        )),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='替换颜色的仪表盘'),
        legend_opts=opts.LegendOpts(is_show=False)
    )
)

c.render_notebook()
```

```python
# 改变其中图中文字的形式

c=(
    Gauge()
    .add(
    '',
        [('完成率',30)],
        title_label_opts=opts.LabelOpts(
            font_size=30,color='red',font_family='Microsoft YaHei'),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='改变轮盘内的字体'))
    
)

c.render_notebook()
```
