```python
from pyecharts import options as opts
from pyecharts.charts import Tree
```

```python
# 导入数据

data = [
    {
        "children": [
            {"name": "B"},
            {
                "children": [{"children": [{"name": "I"}], "name": "E"}, {"name": "F"}],
                "name": "C",
            },
            {
                "children": [
                    {"children": [{"name": "J"}, {"name": "K"}], "name": "G"},
                    {"name": "H"},
                ],
                "name": "D",
            },
        ],
        "name": "A",
    }
]
```

```python
data
```

```python
c=(
    Tree()
    .add('树形图展示',data)
    .set_global_opts(title_opts=opts.TitleOpts(title='基础图形展示'))
)

c.render_notebook()
```

```python
# 上下方向转化

import json
```

```python
# 导入数据
with open('./flare.json','r',encoding='utf-8')as f:
    j=json.load(f)
    
```

```python
# 绘制图形

c=(
    Tree()
    .add(
        "由上向下图形",
        [j],
        collapse_interval=2,
        # 展示的方式
        orient='TB',
        label_opts=opts.LabelOpts(
            position='top',
            horizontal_align='right',
            vertical_align='middle',
            rotate=90
        )
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='由上向下的树图展示'))
)
c.render_notebook()
```

```python
# 换个格式的展示

d=(
    Tree()
    .add(
        '换个方式展示的树图',
        [j],
        collapse_interval=2,
        # 展示的方式
        layout='radial'
        )
    .set_global_opts(title_opts=opts.TitleOpts(title='换个方式展示的树图'))
)

d.render_notebook()
```

```python
# 圆环树图的展示
import asyncio
from aiohttp import TCPConnector, ClientSession

async def get_json_data(url: str) -> dict:
    async with ClientSession(connector=TCPConnector(ssl=False)) as session:
        async with session.get(url=url) as response:
            return await response.json()
        
        
        
# 获取官方的数据
data = await (
    get_json_data(url="https://echarts.apache.org/examples/data/asset/data/flare.json")
)    

data
```

```python
e=(
    Tree(init_opts=opts.InitOpts(width='1400px',height='800px'))
    .add(
        "",
        data=[data],
        pos_top='18%',
        pos_bottom='14%',
        layout='radial',
        symbol='emptyCircle',
        symbol_size=7
    )
    .set_global_opts(tooltip_opts=opts.TooltipOpts(trigger='item',trigger_on='mousemove'))
)

e.render_notebook()
```
