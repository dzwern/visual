```python
# 基础图形的绘制


# 导入模块

import random
from pyecharts import options as opts 
from pyecharts.charts import Bar3D
from pyecharts.faker import Faker
# import datautils
# from datautils import prepare_data
```

```python
# 图形展示

```

```python
# 数据集
data=[(i,j,random.randint(0,12)) for i in range(6) for j in range(24)]

q=(
    # 图形
    Bar3D()
    .add(
    "",
    # 数据
    [[d[1],d[0],d[2]] for d in data],
    xaxis3d_opts=opts.Axis3DOpts(Faker.clock,type_='category'),
    yaxis3d_opts=opts.Axis3DOpts(Faker.week_en,type_='category'),
    zaxis3d_opts=opts.Axis3DOpts(type_='value'),)
    .set_global_opts(
        # 视觉映射配置项，max_指定最大值，右下角的图形展示
        visualmap_opts=opts.VisualMapOpts(max_=20),
        
        title_opts=opts.TitleOpts(title='基础示例'),)
)

q.render_notebook()
```

```python
# 堆叠柱形图

# 导入模块
import random
from pyecharts import options as opts
from pyecharts.charts import Bar3D
```

```python
x_data=y_data=list(range(10))

# 构造data数据集
def genderate_data():
    data=[]
    for j in range(10):
        for k in range(10):
            value=random.randint(0,9)
            data.append([j,k,value*2+4])
    return data
```

```python
# 图形展示
# 图形的另一种写法
bar3D=Bar3D()

for _ in range(10):
    bar3D.add(
    "",
    # 数据集
    genderate_data(),
    shading='lambert',
    xaxis3d_opts=opts.Axis3DOpts(data=x_data,type_='value'),
        yaxis3d_opts=opts.Axis3DOpts(data=y_data,type_='value'),
        zaxis3d_opts=opts.Axis3DOpts(type_='value'),
    )
bar3D.set_global_opts(title_opts=opts.TitleOpts('堆积图展示'))
bar3D.set_series_opts(**{"stack":"stack"})

bar3D.render_notebook()
```
