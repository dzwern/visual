```python
# 组合图形

import pyecharts.options as opts
# 图形模块导入
from pyecharts.charts import Grid, Boxplot, Scatter


y_data = [
 [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880, 1000, 980],
 [960, 940, 960, 940, 880, 800, 850, 880, 900, 840, 830, 790],
 [890, 810, 810, 820, 800, 770, 760, 740, 750, 760, 910, 920],
 [890, 840, 780, 810, 760, 810, 790, 810, 820, 850, 870, 870], 
 [890,840,810,790,810,820,850,870,870,810,740,810],
    ]

scatter_data = [650, 620, 720, 720, 950, 970]
```

```python
# 箱线图
box_plot=Boxplot()
box_plot=(
    box_plot
    # x轴
    .add_xaxis(xaxis_data=['1','2','3','4','5'])
    # y轴数据
    .add_yaxis(series_name='',y_axis=box_plot.prepare_data(y_data))
    # 全局配置
    .set_global_opts(
        # 标题配置
        title_opts=opts.TitleOpts(
            pos_left='center',title='箱线图'),
        # 工具配置  
        tooltip_opts=opts.TooltipOpts(trigger='item',axis_pointer_type='shadow'),
        # x轴线配置
        xaxis_opts=opts.AxisOpts(
            type_='category',
            # 是否有间距
            boundary_gap=True,
            # 分割区域配置项
            splitarea_opts=opts.SplitAreaOpts(is_show=False),
            # x轴标签绘制
            axislabel_opts=opts.LabelOpts(formatter='expr{value}'),
            # 分割线配置项
            splitline_opts=opts.SplitLineOpts(is_show=False),
        ),
        #y轴线配置
        yaxis_opts=opts.AxisOpts(
            type_='value',
            name='km/s minus 290000',
            # 分割区域配置项
            splitarea_opts=opts.SplitAreaOpts(is_show=True,
                                # 分割区域的样式配置项
                                areastyle_opts=opts.AreaStyleOpts(opacity=1)),
            ),)
    .set_series_opts(tooltip_opts=opts.TooltipOpts(formatter='{b}:{c}')))

box_plot.render_notebook()
```

```python
# 散点图的绘制
scatter=(
    Scatter()
    .add_xaxis(xaxis_data=['expr1','expr2','expr3','expr4','expr5'])
    .add_yaxis(series_name='',y_axis=scatter_data)
    .set_global_opts(
        # 标题配置（位置调整）
        title_opts=opts.TitleOpts(
            pos_left='10%',
            pos_top='93%',
            title='upper:Q3+1.5*IQR \nlower:Q1-1.5*IQR',
            # 文字样式的配置
            title_textstyle_opts=opts.TextStyleOpts(
                border_color='#999',border_width=1,font_size=14),),
        # y轴配置
        yaxis_opts=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(is_show=False),
            axistick_opts=opts.AxisTickOpts(is_show=False),),
    ))
scatter.render_notebook()
```

```python
# 组合图

grid=(
    # 画布大小
    Grid(init_opts=opts.InitOpts(width='800px',height='500px'))
    .add(
        box_plot,
        grid_opts=opts.GridOpts(pos_left='10%',pos_right='10%',
                                pos_bottom='15%'),
    )
    .add(
        scatter,
        grid_opts=opts.GridOpts(pos_left='10%',pos_right='10%',
                                pos_bottom='15%'),
    ))
    
grid.render_notebook()
```
