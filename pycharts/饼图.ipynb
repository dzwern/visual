```python
# 配置项
from pyecharts import options as opts
# 饼图
from pyecharts.charts import Pie
# 数据选择
from pyecharts.faker import Faker

```

```python
c=(
    Pie()
    .add("",[list(z)for z in zip(Faker.choose(),Faker.values())])
    .set_global_opts(title_opts=opts.TitleOpts(title='饼图')
                     # 对图例进行更改
                    ,legend_opts=opts.LegendOpts(pos_left='15%'))
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{c}'))

)

c.render_notebook()
```

```python
# 更改颜色
d=(
    Pie()
    .add("",[list(z) for z in zip(Faker.choose(),Faker.values())])
    .set_colors(['blue','green','yellow','red','pink','orange','purple'])
    # 全局设置
    .set_global_opts(title_opts=opts.TitleOpts(title='颜色更改饼图'))
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{c}'))
)

d.render_notebook()
```

```python
# 圆环图
e=(
    Pie()
    .add("",[list(z) for z in zip(Faker.choose(),Faker.values())],
         radius=['%40','55%'],
         # 标签的设置
         label_opts=opts.LabelOpts(
             position='outside',
             formatter='{a|{a}}{abg|}\n{hr|}\n{b|{b}: }{c} {per|{d}%} ',
             background_color='#eee',
             border_color='#aaa',
             border_width=1,
             border_radius=4,
             rich={ 
                 "a": {"color": "#999", "lineHeight": 22, "align": "center"},
                "abg": {
                    "backgroundColor": "#e3e3e3",
                    "width": "100%",
                    "align": "right",
                    "height": 22,
                    "borderRadius": [4, 4, 0, 0],
                },
                "hr": {
                    "borderColor": "#aaa",
                    "width": "100%",
                    "borderWidth": 0.5,
                    "height": 0,
                },
                "b": {"fontSize": 16, "lineHeight": 33},
                "per": {
                    "color": "#eee",
                    "backgroundColor": "#334455",
                    "padding": [2, 4],
                    "borderRadius": 2,
                },
            },
         ),
        )
    .set_global_opts(title_opts=opts.TitleOpts(title='标注修改图形'))
         
)
e.render_notebook()
```

```python
f=(
    Pie(init_opts=opts.InitOpts(width='500px',height='500px'))
    .add("",[list(z) for z in zip(Faker.choose(),Faker.values())]
        ,radius=["50%",'70%'],
#         label_opts=opts.LabelOpts(is_show=True,position='center')
        )
    .set_global_opts(legend_opts=opts.LegendOpts(pos_left='left',orient="vertical"))
    .set_series_opts(
        tooltip_opts=opts.TooltipOpts(
        trigger='item',formatter='{a}<br/>{b}: {c} ({d}%)')
    )
)

f.render_notebook()
```

```python
# 多个图形

from pyecharts.commons.utils import JsCode


fn="""
    function(params){
    if (params.name=='其他')
            return '\\n\\n\\n' + params.name + ' : ' + params.value + '%';
        return params.name + ' : ' + params.value + '%';
    }

"""

# 标签函数
def new_label_opts():
    return opts.LabelOpts(formatter=JsCode(fn),position='center')



h=(
    Pie()
    .add("",[list(z) for z in zip(['剧情','其他'],[25,75])],
         # 位置
         center=['20%','30%'],
         radius=[60,80],
         label_opts=new_label_opts(),
        )
    .add("",[list(z) for z in zip(['奇幻','其他'],[24,76])],
         # 位置
         center=['55%','30%'],
         radius=[60,80],
         label_opts=new_label_opts(),
        )
    .add("",[list(z) for z in zip(['爱情','其他'],[14,86])],
         # 位置
         center=['20%','70%'],
         radius=[60,80],
         label_opts=new_label_opts(),
        )
    .add("",[list(z) for z in zip(['动作','其他'],[11,89])],
         # 位置
         center=['55%','70%'],
         radius=[60,80],
         label_opts=new_label_opts(),
        ) 
    .set_global_opts(
        title_opts=opts.TitleOpts(title='多图展示'),
        legend_opts=opts.LegendOpts(
            type_='scroll',pos_top='20%',pos_left='80%',orient='vertical'))
)

h.render_notebook()
```

```python
# 玫瑰图
i=(
    Pie(init_opts=opts.InitOpts(width='1200px',height='500px'))
    .add("",[list(z) for z in zip(Faker.choose(),Faker.values())],
         radius=['30%','75%'],
         center=['25%','50%'],
         rosetype='radius',
         label_opts=opts.LabelOpts(is_show=True),
        )
    .add(
        "",[list(z) for z in zip(Faker.choose(),Faker.values())],
        radius=['30%','75%'],
        center=['75%','50%'],
        rosetype='area'
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='玫瑰图展示'),
                            legend_opts=opts.LegendOpts(
            pos_left='50%',orient='vertical'))
                    )
    
i.render('./玫瑰图展示.html')

# i.render_notebook()
```
