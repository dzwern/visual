```python
# 多个图形展示

import re
import asyncio
from aiohttp import TCPConnector,ClientSession

import pyecharts.options as opts
from pyecharts.charts import TreeMap
```

```python
# 导入数据
async def get_json_data(url):
    async with ClientSession(connector=TCPConnector(ssl=False)) as session:
        async with session.get(url=url)as response:
            return await response.json()
```

```python
data = await(
    get_json_data(
        url="https://echarts.apache.org/examples/data/asset/data/"
        "ec-option-doc-statistics-201604.json"
    )
)
data
```

```python
tree_map_data:dict={"children":[]}
    
def convert(source,target,base_path):
    for key in source:
        if base_path != "":
            path = base_path + "." + key
        else:
            path = key
        if re.match(r"/^\$/", key):
            pass
        else:
            child = {"name": path, "children": []}
            target["children"].append(child)
            if isinstance(source[key], dict):
                convert(source[key], child, path)
            else:
                target["value"] = source["$count"]
                
convert(source=data,target=tree_map_data,base_path="")
```

```python
d=(
    TreeMap(init_opts=opts.InitOpts(width='800px',height='720px'))
    .add(
        series_name='option',
        data=tree_map_data['children'],
        visual_min=300,
        leaf_depth=1,
        label_opts=opts.LabelOpts(position='inside')
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(is_show=False),
        title_opts=opts.TitleOpts(title='树型图的绘制',subtitle="绘制图形",
                                  pos_left="leafDepth")
    )
)

d.render_notebook()
```
