```python
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker
from pyecharts.charts import Geo
from pyecharts.charts import BMap
```

```python
'''
Map:负责的是地图的模块
Geo：负责的是地理坐标系
BMap:负责的是百度地图模块
'''

```

```python
c = (
    Map()
    .add("中国地图分布", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
    .set_global_opts(title_opts=opts.TitleOpts(title="Map-基本示例"))

)

c.render_notebook()
```

```python
# 世界地图
c = (
    Map()
    .add("世界地图", [list(z) for z in zip(Faker.country, Faker.values())], "world")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-世界地图"),
        visualmap_opts=opts.VisualMapOpts(max_=200),
    )
)

c.render_notebook()
```

```python
# 城市图
c = (
    Map()
    .add(
        "中国城市分布",
        [list(z) for z in zip(Faker.guangdong_city, Faker.values())],
        "china-cities",
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-中国地图（带城市）"),
        visualmap_opts=opts.VisualMapOpts(),
    )
)
c.render_notebook()
```

```python
# 图例时连续型
c = (
    Map()
    .add("中国地图分布-连续型", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-VisualMap（连续型）"),
        # 连续型，分段型，分段型使用参数is_piecewise
        visualmap_opts=opts.VisualMapOpts(max_=200),
    )
)
c.render_notebook()
```

```python
c = (
    Map()
    .add("中国地图分布-分段型", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-VisualMap（分段型）"),
        # 连续型，分段型，分段型使用参数is_piecewise
        visualmap_opts=opts.VisualMapOpts(max_=200,split_number=5,is_piecewise=True),
    )
)
c.render_notebook()
```

```python
c = (
    Map()
    .add("广东地图分布", [list(z) for z in zip(Faker.guangdong_city, Faker.values())], "广东")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-广东地图"), 
        visualmap_opts=opts.VisualMapOpts()
    )
)
c.render_notebook()
```

```python
c = (
        Map()
        .add("河南地图分布", [list(z) for z in zip(['郑州市','开封市','洛阳市','商丘市','平顶山市'], [22,100,27,53,42])], "河南")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-河南地图"),
            visualmap_opts=opts.VisualMapOpts(max_=200),
        )
    )
c.render_notebook()
```

```python
# Geo绘制图形

c=(
    Geo()
    .add_schema(maptype='china')
    .add('中国地图',[list(z) for z in zip(Faker.provinces,Faker.values())])
    # 标签配置
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    # 全局配置
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(),
                     title_opts=opts.TitleOpts('Geo-基本示例'))

)

c.render_notebook()
```

```python
# Geo绘制图形,热力图
from pyecharts.globals import ChartType

c=(
    Geo()
    .add_schema(maptype='china')
    .add('中国地图',[list(z) for z in zip(Faker.provinces,Faker.values())],
        type_=ChartType.HEATMAP)
    # 标签配置
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    # 全局配置
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(),
                     title_opts=opts.TitleOpts('Geo-基本示例'))

)

c.render_notebook()
```

```python
# Geo绘制图形,热力图
from pyecharts.globals import ChartType

c=(
    Geo()
    .add_schema(maptype='china',
               itemstyle_opts=opts.ItemStyleOpts(color="#323c48", border_color="#111"),)
    .add('中国地图',[list(z) for z in zip(Faker.provinces,Faker.values())],
        type_=ChartType.HEATMAP)
    # 标签配置
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    # 全局配置
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(),
                     title_opts=opts.TitleOpts('Geo-基本示例'))

)

c.render_notebook()
```

```python
# 绘制省份图
c = (
        Geo()
        .add_schema(maptype="河南")
        .add(
            "河南地图分布",
            [list(z) for z in zip(['郑州市','商丘市','安阳市','信阳市','漯河市'], [150,100,300,200,500])],
            # 涟漪效果
            type_=ChartType.EFFECT_SCATTER,
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(),
            title_opts=opts.TitleOpts(title="Geo-河南省份地图"),
        )
    )

c.render_notebook()
```

```python
from pyecharts.globals import SymbolType

c = (
        Geo()
        .add_schema(maptype="china")
        # 涟漪效果图
        .add(
            "",
            [("深圳", 120), ("哈尔滨", 66), ("河南", 77), ("重庆", 88), ("上海", 100), ("乌鲁木齐", 30),("北京", 30),("武汉",70)],
            type_=ChartType.EFFECT_SCATTER,
            color="green",
        )
        # 流转线图
        .add(
            "地图流转图",
            [("北京", "上海"), ("武汉", "深圳"),("重庆", "河南"),("哈尔滨", "重庆"),("乌鲁木齐", "哈尔滨"),("深圳", "乌鲁木齐"),("武汉", "北京")],
            type_=ChartType.LINES,
            
            effect_opts=opts.EffectOpts(
                symbol=SymbolType.ARROW, symbol_size=6, color="blue"
            ),
            linestyle_opts=opts.LineStyleOpts(curve=0.2),
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="地图流转图"))
    )

c.render_notebook()
```

```python
# 绘制百度地图

from pyecharts import options as opts
from pyecharts.charts import BMap
from pyecharts.faker import Faker

c = (
    BMap()
    .add_schema(baidu_ak="FAKE_DZW", center=[120.13066322374, 30.240018034923])
    .add(
        "bmap",
        [list(z) for z in zip(Faker.provinces, Faker.values())],
        label_opts=opts.LabelOpts(formatter="{b}"),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="BMap-基本示例"))
    
)

c.render_notebook()
```
